```python
import math
```

```python
# Constants
math.pi                  # returns the value of pi
math.e                   # returns the value of e (the base of the natural logarithm)
```

```python
# Basic functions
x,y = 1,20

math.sqrt(x)             # returns the square root of x
math.exp(x)              # returns e raised to the power of x
math.log(x)              # returns the natural logarithm of x (base e)
math.log10(x)            # returns the base-10 logarithm of x
math.pow(x, y)           # returns x raised to the power of y
math.radians(x)          # converts x from degrees to radians
math.degrees(x)          # converts x from radians to degrees
```

```python
# Trigonometric functions
math.sin(x)              # returns the sine of x (in radians)
math.cos(x)              # returns the cosine of x (in radians)
math.tan(x)              # returns the tangent of x (in radians)
math.asin(x)             # returns the arcsine of x (in radians)
math.acos(x)             # returns the arccosine of x (in radians)
math.atan(x)             # returns the arctangent of x (in radians)
```

```python
# Other functions
math.ceil(x)             # returns the ceiling of x (i.e., the smallest integer greater than or equal to x)
math.floor(x)            # returns the floor of x (i.e., the largest integer less than or equal to x)
math.fabs(x)             # returns the absolute value of x
math.factorial(x)        # returns the factorial of x
math.gcd(x, y)           # returns the greatest common divisor of x and y
math.isfinite(x)         # returns True if x is a finite number, False otherwise
math.isnan(x)            # returns True if x is not a number (NaN), False otherwise
math.modf(x)             # returns the fractional and integer parts of x as a tuple
math.trunc(x)            # returns the truncated value of x (i.e., the integer part of x)
```
