The standard deviation tells us how much the data points deviate from the mean or average value of the data set. It is calculated as the square root of the variance of the data set.

```python
np.var(array_2)
```

Intuitively, the variance measures how much the data points deviate from their average value. A high variance indicates that the data points are spread out over a wide range of values, while a low variance indicates that the data points are tightly clustered around their mean value.

```python
# demon of std and var
high_var_array = np.array([1,100,200,300,4000,5000])
low_var_array = np.array([2,4,6,8,10])
```

```python
np.var(high_var_array), np.var(low_var_array)
```

```python
np.std(high_var_array), np.std(low_var_array)
```

```python
np.mean(high_var_array), np.mean(low_var_array)
```

# Dot product vs element wise

In NumPy, the dot product is a mathematical operation between two arrays that produces a scalar value. It is defined as the sum of the products of the corresponding elements of the two arrays.

The element-wise multiplication in NumPy is an operation where each element of one array is multiplied by the corresponding element of another array to produce a new array of the same shape as the original arrays.

```python
np.random.seed(42)

mat1 = np.random.randint(10,size = (5,3))
mat2 = np.random.randint(10,size = (5,3))

mat1, mat2
```

```python
# element wise multiplication
mat1 * mat2
```

![dot_product](img/dot_product.png)


```python
# dot multiplication
#np.dot(mat1, mat2) # shape error
mat1, mat2.T, np.dot(mat1,mat2.T), mat1.shape, mat2.T.shape, np.dot(mat1,mat2.T).shape
```

## Dot product application

```python
np.random.seed(42)
sales_amounts=np.random.randint(20, size=(5,3))
sales_amounts
```

```python
# create dataframe
weekly_sales = pd.DataFrame(sales_amounts, index=["mon", 'tue', 'wen', 'thur', 'fri'], columns=['amonds', 'peanuts', 'cashew'])
weekly_sales
```

```python
#create prices
prices = np.array([10,8,12])
```

```python
#create prices dataframe
butter_prices = pd.DataFrame(prices.reshape(1,3), index=['price'], columns=['amonds', 'peanuts', 'cashew'])
butter_prices
```

```python
total_sales = prices.dot(sales_amounts.T)
total_sales
```

```python
# create daily_sales
daily_sales = butter_prices.dot(weekly_sales.T)
daily_sales
```

```python
weekly_sales['Total ($)'] = daily_sales.T
weekly_sales
```
